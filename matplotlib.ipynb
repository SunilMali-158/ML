```python
x = np.linspace(1,10,100) 
```

```python
print(x)
```

```python
sin = np.sin(x)
```

```python
plt.plot(x,sin)
```

```python
cos = np.cos(x)
```

```python
plt.plot(x,cos)
```

```python
plt.plot(cos,sin)
```

```python
sin2 = np.sin(x)**2
```

```python
plt.plot(sin, label= "sin function")
plt.plot(cos, label= "cos function")
plt.plot(sin2,label = "sinsqure function")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
```
